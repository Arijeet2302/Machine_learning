```python
import pandas as pd 

data = pd.read_csv("House_Rent_Dataset.csv")
data.head()
```

```python
data.drop(['Posted On', 'Area Locality', 'Floor'], axis=1, inplace=True)
```

```python
data.head()
data['Rent'].describe().astype(int)
```

```python
data['City'].value_counts()
```

```python
from sklearn.compose import ColumnTransformer
from  sklearn.preprocessing import OneHotEncoder

encode_target = [ 'City', 'Area Type', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact']

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), encode_target)], remainder='passthrough')

new_data = ct.fit_transform(data)

new_data = pd.DataFrame(new_data)

new_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
X = new_data.drop([19], axis = 1)
y = new_data[19]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
x_train
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train.iloc[:, 18:] = sc.fit_transform(x_train.iloc[:, 18:])
x_test.iloc[:, 18:] = sc.transform(x_test.iloc[:, 18:] )
```

```python
dst = DecisionTreeRegressor()

dst.fit(x_train, y_train)

y_pred = dst.predict(x_test)
y_pred[0]
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

acc = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(mae)
print(mse)
print(acc)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

r2_score(y_test, y_pred)
```
